```python
# 실습 #1

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
result_max = []
result_min = []
for row in data :
    if row[-1] != '' and row[-2] != '' :
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '19' :
            result_max.append(float(row[-1]))
            result_min.append(float(row[-2]))

plt.plot(result_max)
plt.plot(result_min)
plt.show()
```

```python
# 실습 #2

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
jan = []
aug = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '01' :
            jan.append(float(row[-1]))
        elif row[0].split('-')[1] == '08' :
            aug.append(float(row[-1]))
plt.hist(jan, bins=100, color='b')
plt.hist(aug, bins=100, color='r')
plt.show()
```
